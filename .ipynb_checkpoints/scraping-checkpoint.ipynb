<h1>Scraping Data</h1>

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
chrome_service = Service(executable_path=r"YOUR_PATH\\chromedriver.exe")
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
options.add_experimental_option('excludeSwitches', ['enable-logging'])
options.add_argument(f'user-agent={user_agent}')

driver = webdriver.Chrome(service=chrome_service, options=options)
url = "https://vinacorp.vn"
driver.get(url)
```

```python
def get_name_company(symbol):
    url = "https://vinacorp.vn/trade-history/" + symbol
    driver.get(url)
    company_name = driver.find_element(By.XPATH, '//*[@id="st-container"]/div/div/div/div/section[4]/div/div/div[1]/div/a[2]/span').text
    return company_name

list_company_name = []
for i in range(0, len(symbol),1):
    list_company_name.append(get_name_company(symbol[i]))

print(list_company_name)
```

```python
def stock_historical_price(symbol):
    d_start = "2020-01-01"
    d_end = "2022-12-15"
    count = datetime.strptime(d_end, "%Y-%m-%d") - datetime.strptime(d_start, "%Y-%m-%d")
    url =  "https://finfo-api.vndirect.com.vn/v4/stock_prices?sort=date&q=code:"+ symbol + "~date:gte:" + d_start + "~date:lte:"+ d_end + "&size=" + str(count.days);                                                                                                         session = requests.Session()
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"}
    session = requests.Session()
    session.cookies.get_dict()
    response = session.get(url,headers = headers)
    data = json.loads(response.content)
    df = json_normalize(data.get('data'))
    return df
```

**Dữ liệu tính tới ngày 15/12/2022**

```python
listDict = []
for i in range(0, len(symbol),1):
    df = stock_historical_price(symbol[i])
    for data in df.index:
        info_dict = {'Stock': symbol[i], 'Company': list_company_name[i] ,'Floor': df['floor'][data], 'Date': df['date'][data], 'ClosingPrice': df['close'][data]}
        listDict.append(info_dict)
```

```python
df = pd.DataFrame.from_dict(listDict)
df
```

```python
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
df
```

```python
df = df[df['Floor'] == "HOSE"]
df
```

```python
df = df.pivot(values="ClosingPrice", index=["Stock", "Company", "Floor"], columns='Date')
df
```

```python
df.to_csv(r'History Closing Price.csv', index = True, header=True)
```
